```python
def multi(v1, v2) :
    retList = []
    res1 = v1 + v2
    res2 = v1 - v2
    res3 = v1 * v2
    res4 = v1 // v2
    res5 = v1 % v2
    res6 = v1 ** v2
    retList.append(res1)
    retList.append(res2)
    retList.append(res3)
    retList.append(res4)
    retList.append(res5)
    retList.append(res6)
    return retList
    
myList = []
a, b, c, d, e, f = 0, 0, 0, 0, 0, 0

myList = multi(100, 20)
a = myList[0]
b = myList[1]
c = myList[2]
d = myList[3]
e = myList[4]
f = myList[5]
print("multi()에서 반환한 값 ==> ", myList)
```
